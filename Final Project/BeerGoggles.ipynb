```python
# Import Pandas
import pandas as pd

# Load Movies Metadata
metadata = pd.read_csv('cleanedfinalbeerdata.csv', low_memory=False)

# Print the first three rows
metadata.head(3)
```

```python
C = metadata['abv'].mean()
print(C)
```

```python
metadata['ibu'] = metadata['ibu'].fillna('')
```

```python
metadata['srmId'] = metadata['srmId'].fillna('')
```

```python
metadata.head(10)
```

```python
bins = [2.5, 4, 6, 8, 10, 16]
labels = [0,1,2,3,4]
metadata['abv_category'] = pd.cut(metadata['abv'], bins=bins, labels=labels, include_lowest=True)
display(metadata.head(n=2))
```

```python
bins = [1, 20, 50, 75, 100, 150, 200, 700]
labels = [0,1,2,3,4,5,6]
metadata['ibu_category'] = pd.cut(metadata['ibu'], bins=bins, labels=labels, include_lowest=True)
display(metadata.head(n=2))
```

```python
bins = [1, 2, 5, 10, 20, 30, 41]
labels = [0,1,2,3,4,5]
metadata['srm_category'] = pd.cut(metadata['srmId'], bins=bins, labels=labels, include_lowest=True)
display(metadata.head(n=2))
```
